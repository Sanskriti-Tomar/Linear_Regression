```python
df2_AP_rel = df2_AP[ ['HHID', 'Item_Code','Last_365days_Quantity', 'Last_365days_Value','Last_30days_Quantity', 	'Last_30days_Value'] ]
```

```python
clothing_bedding_and_footwear_quantity  = df2_AP_rel.pivot(index=['HHID'], columns=['Item_Code'],values=['Last_365days_Quantity']).reset_index()
clothing_bedding_and_footwear_value  = df2_AP_rel.pivot(index=['HHID'], columns=['Item_Code'],values=['Last_365days_Value']).reset_index()
```

```python
clothing_bedding_and_footwear_quantity[("Last_365days_Quantity", "352")] = clothing_bedding_and_footwear_quantity[("Last_365days_Quantity", "352")]/200
```

```python
clothing_bedding_and_footwear_quantity[("Last_365days_Quantity", "353")] = clothing_bedding_and_footwear_quantity[("Last_365days_Quantity", "353")]/200
```

```python
clothing_bedding_and_footwear_quantity[("Last_365days_Quantity", "383")] = clothing_bedding_and_footwear_quantity[("Last_365days_Quantity", "383")]/250
```

```python
clothing_bedding_and_footwear_quantity.drop([('Last_365days_Quantity', '374'),('Last_365days_Quantity', '379'), ('Last_365days_Quantity', '389'),('Last_365days_Quantity', '390'), ('Last_365days_Quantity', '391'),  ('Last_365days_Quantity', '392'), ('Last_365days_Quantity', '393'), ('Last_365days_Quantity', '394'), ('Last_365days_Quantity', '395')], axis = 1, inplace = True)
```

```python
clothing_bedding_and_footwear_value.drop([('Last_365days_Value', '374'),('Last_365days_Value', '379'), ('Last_365days_Value', '389'),('Last_365days_Value', '390'), ('Last_365days_Value', '391'),  ('Last_365days_Value', '392'), ('Last_365days_Value', '393'), ('Last_365days_Value', '394'), ('Last_365days_Value', '395')], axis = 1, inplace = True)
```

```python
clothing_bedding_and_footwear_quantity
```

```python
clothing_bedding_and_footwear_value
```

```python
clothing_bedding_and_footwear_quantity["Total_Quantity"]= clothing_bedding_and_footwear_quantity.sum(axis=1)
clothing_bedding_and_footwear_value["Total_Value"]= clothing_bedding_and_footwear_value.sum(axis=1)
```

```python
#clothing_bedding_and_footwear_quantity.fillna(0,inplace=True)
#clothing_bedding_and_footwear_value.fillna(0, inplace=True)
```

```python
clothing_bedding_and_footwear_Total = pd.merge(clothing_bedding_and_footwear_quantity,clothing_bedding_and_footwear_value, how = "inner" , on ="HHID" )
clothing_bedding_and_footwear_Total
```

```python
AGG_clothing_bedding_and_footwear = clothing_bedding_and_footwear_Total[['HHID','Total_Quantity','Total_Value']]
```

```python
AGG_clothing_bedding_and_footwear.rename(columns={"Total_Quantity":"CLOTHING_BEDDING_AND_FOOTWEAR_TOTAL_QUANTITY","Total_Value" :"CLOTHING_BEDDING_AND_FOOTWEAR_TOTAL_VALUE"}, inplace = True)
```

```python
AGG_clothing_bedding_and_footwear["PRICE"] = AGG_clothing_bedding_and_footwear["CLOTHING_BEDDING_AND_FOOTWEAR_TOTAL_VALUE"]/AGG_clothing_bedding_and_footwear["CLOTHING_BEDDING_AND_FOOTWEAR_TOTAL_QUANTITY"]
AGG_clothing_bedding_and_footwear.set_index('HHID', inplace= True)
```
